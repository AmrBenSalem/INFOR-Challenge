```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 500)
```

```python
train = pd.read_csv("training_set.csv")
test = pd.read_csv("testing_set.csv")
hotels_capacities = pd.read_csv("hotels_capacities.csv",sep=";")
```

```python
train["Date"] = pd.to_datetime(train["Date"],dayfirst=True)
test["Date"] = pd.to_datetime(test["Date"],dayfirst=True)
```

```python
train["Tot_BKG"] = train.BKG_0+train.BKG_1+train.BKG_14+train.BKG_21+train.BKG_28+train.BKG_7
train["Tot_BKG_mean"] = (train.BKG_0+train.BKG_1+train.BKG_14+train.BKG_21+train.BKG_28+train.BKG_7)/6

train["Tot_Revenue"] = train.Revenue_0+train.Revenue_1+train.Revenue_14+train.Revenue_21+train.Revenue_28+train.Revenue_7
train["Tot_Revenue_mean"] = (train.Revenue_0+train.Revenue_1+train.Revenue_14+train.Revenue_21+train.Revenue_28+train.Revenue_7)/6

train["Tot_HotelPrice"] = train.HotelPrice_14 + train.HotelPrice_21 + train.HotelPrice_28
train["Tot_HotelPrice_mean"] = (train.HotelPrice_14 + train.HotelPrice_21 + train.HotelPrice_28)/3

train["Tot_PriceComp0"] = train.PriceComp1_0 + train.PriceComp2_0 + train.PriceComp3_0 + train.PriceComp4_0
train["Tot_PriceComp0_mean"] = (train.PriceComp1_0 + train.PriceComp2_0 + train.PriceComp3_0 + train.PriceComp4_0)/4

train["Tot_PriceComp14"] = train.PriceComp1_14 + train.PriceComp2_14 + train.PriceComp3_14 + train.PriceComp4_14
train["Tot_PriceComp14_mean"] = (train.PriceComp1_14 + train.PriceComp2_14 + train.PriceComp3_14 + train.PriceComp4_14)/4

train["Tot_PriceComp21"] = train.PriceComp1_21 + train.PriceComp2_21 + train.PriceComp3_21 + train.PriceComp4_21
train["Tot_PriceComp21_mean"] = (train.PriceComp1_21 + train.PriceComp2_21 + train.PriceComp3_21 + train.PriceComp4_21)/4

train["Tot_PriceComp28"] = train.PriceComp1_28 + train.PriceComp2_28 + train.PriceComp3_28 + train.PriceComp4_28
train["Tot_PriceComp28_mean"] = (train.PriceComp1_28 + train.PriceComp2_28 + train.PriceComp3_28 + train.PriceComp4_28)/4
```

```python
test["Tot_BKG"] = test.BKG_0+test.BKG_1+test.BKG_14+test.BKG_21+test.BKG_28+test.BKG_7
test["Tot_BKG_mean"] = (test.BKG_0+test.BKG_1+test.BKG_14+test.BKG_21+test.BKG_28+test.BKG_7)/6

test["Tot_Revenue"] = test.Revenue_0+test.Revenue_1+test.Revenue_14+test.Revenue_21+test.Revenue_28+test.Revenue_7
test["Tot_Revenue_mean"] = (test.Revenue_0+test.Revenue_1+test.Revenue_14+test.Revenue_21+test.Revenue_28+test.Revenue_7)/6

test["Tot_HotelPrice"] = test.HotelPrice_14 + test.HotelPrice_21 + test.HotelPrice_28
test["Tot_HotelPrice_mean"] = (test.HotelPrice_14 + test.HotelPrice_21 + test.HotelPrice_28)/3

test["Tot_PriceComp0"] = test.PriceComp1_0 + test.PriceComp2_0 + test.PriceComp3_0 + test.PriceComp4_0
test["Tot_PriceComp0_mean"] = (test.PriceComp1_0 + test.PriceComp2_0 + test.PriceComp3_0 + test.PriceComp4_0)/4

test["Tot_PriceComp14"] = test.PriceComp1_14 + test.PriceComp2_14 + test.PriceComp3_14 + test.PriceComp4_14
test["Tot_PriceComp14_mean"] = (test.PriceComp1_14 + test.PriceComp2_14 + test.PriceComp3_14 + test.PriceComp4_14)/4

test["Tot_PriceComp21"] = test.PriceComp1_21 + test.PriceComp2_21 + test.PriceComp3_21 + test.PriceComp4_21
test["Tot_PriceComp21_mean"] = (test.PriceComp1_21 + test.PriceComp2_21 + test.PriceComp3_21 + test.PriceComp4_21)/4

test["Tot_PriceComp28"] = test.PriceComp1_28 + test.PriceComp2_28 + test.PriceComp3_28 + test.PriceComp4_28
test["Tot_PriceComp28_mean"] = (test.PriceComp1_28 + test.PriceComp2_28 + test.PriceComp3_28 + test.PriceComp4_28)/4
```

```python
train["year"] = train.Date.dt.year
train["month"] = train.Date.dt.month
train["day"] = train.Date.dt.day
train["week"] = train.Date.dt.week
train["dayoftheweek"] = train.Date.dt.dayofweek
```

```python
test["year"] = test.Date.dt.year
test["month"] = test.Date.dt.month
test["day"] = test.Date.dt.day
test["week"] = test.Date.dt.week
test["dayoftheweek"] = test.Date.dt.dayofweek
```

```python
train["mean_enc_target"] = train["Hotel_id"].map(train.groupby("Hotel_id")["HotelPrice_0"].mean())
train["std_enc_target"] = train["Hotel_id"].map(train.groupby("Hotel_id")["HotelPrice_0"].std())

train["mean_enc_target_weekday"] = train.merge(train.groupby(["Hotel_id","dayoftheweek"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","dayoftheweek"],how="left")["HotelPrice_0_y"]
train["std_enc_target_weekday"] = train.merge(train.groupby(["Hotel_id","dayoftheweek"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","dayoftheweek"],how="left")["HotelPrice_0_y"]

train["mean_enc_target_month"] = train.merge(train.groupby(["Hotel_id","month"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","month"],how="left")["HotelPrice_0_y"]
train["std_enc_target_month"] = train.merge(train.groupby(["Hotel_id","month"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","month"],how="left")["HotelPrice_0_y"]

train["mean_enc_target_year"] = train.merge(train.groupby(["Hotel_id","year"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","year"],how="left")["HotelPrice_0_y"]
train["std_enc_target_year"] = train.merge(train.groupby(["Hotel_id","year"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","year"],how="left")["HotelPrice_0_y"]

train["mean_enc_target_week"] = train.merge(train.groupby(["Hotel_id","week"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","week"],how="left")["HotelPrice_0_y"]
train["std_enc_target_week"] = train.merge(train.groupby(["Hotel_id","week"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","week"],how="left")["HotelPrice_0_y"]
```

```python
test["mean_enc_target"] = test["Hotel_id"].map(train.groupby("Hotel_id")["HotelPrice_0"].mean())
test["std_enc_target"] = test["Hotel_id"].map(train.groupby("Hotel_id")["HotelPrice_0"].std())

test["mean_enc_target_weekday"] = test.merge(train.groupby(["Hotel_id","dayoftheweek"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","dayoftheweek"],how="left")["HotelPrice_0"]
test["std_enc_target_weekday"] = test.merge(train.groupby(["Hotel_id","dayoftheweek"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","dayoftheweek"],how="left")["HotelPrice_0"]

test["mean_enc_target_month"] = test.merge(train.groupby(["Hotel_id","month"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","month"],how="left")["HotelPrice_0"]
test["std_enc_target_month"] = test.merge(train.groupby(["Hotel_id","month"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","month"],how="left")["HotelPrice_0"]

test["mean_enc_target_year"] = test.merge(train.groupby(["Hotel_id","year"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","year"],how="left")["HotelPrice_0"]
test["std_enc_target_year"] = test.merge(train.groupby(["Hotel_id","year"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","year"],how="left")["HotelPrice_0"]

test["mean_enc_target_week"] = test.merge(train.groupby(["Hotel_id","week"])["HotelPrice_0"].mean().reset_index(),on=["Hotel_id","week"],how="left")["HotelPrice_0"]
test["std_enc_target_week"] = test.merge(train.groupby(["Hotel_id","week"])["HotelPrice_0"].std().reset_index(),on=["Hotel_id","week"],how="left")["HotelPrice_0"]
```

```python
train["Hotel_id"] = train["Hotel_id"].astype("str")
test["Hotel_id"] = test["Hotel_id"].astype("str")
```

```python
hotels_capacities["Hotel"] = hotels_capacities["Hotel"].astype(str)
```

```python
train = train.merge(hotels_capacities,how="left",left_on="Hotel_id",right_on="Hotel").drop("Hotel",axis=1)
```

```python
test = test.merge(hotels_capacities,how="left",left_on="Hotel_id",right_on="Hotel").drop("Hotel",axis=1)
```

```python
X = train.drop(["Row_id","Date","Date_id","HotelPrice_0"],axis=1)
y = train.HotelPrice_0
```

```python
X["strat"] = X.Hotel_id.astype(str)+"_"+X.year.astype(str)+"_" +X.month.astype(str)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=0,stratify=X.strat)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
ln = LinearRegression()
ln.fit(X_train,y_train)
y_pred = ln.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
y_pred = ln.predict(test.drop(["Row_id","Date","Date_id"],axis=1))
c = pd.DataFrame({"Row_id":test.Row_id,"HotelPrice_0":y_pred})
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn import datasets
```

```python
import lightgbm as lgb
```

```python
iris = datasets.load_iris()
```

```python
irisdd = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
model_lgb = lgb.LGBMClassifier()
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
sk = StratifiedKFold()
```

```python
gridParams = {
    'learning_rate': [0.05,0.006,0.007],
    'n_estimators': [40],
    'num_leaves': [6,8,12,16],
    'boosting_type' : ['gbdt'],
    'objective' : ['binary'],
    'colsample_bytree' : [0.65, 0.66],
    'subsample' : [0.7,0.75],
    'reg_alpha' : [1,1.2],
    'reg_lambda' : [1,1.2,1.4],
    }
```

```python
clf = GridSearchCV(model_lgb,gridParams,verbose=1,cv=sk.split(irisdd,iris.target))
```

```python
clf.fit(irisdd,iris.target)
```

```python
clf.best_score_
```
